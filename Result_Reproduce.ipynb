**Numpy**

```python
import numpy as np

# Create 1D array (10 to 29)
arr = np.arange(10, 30)
print('1D Array:', arr)

# Reshape to 5x4 matrix
matrix = arr.reshape(5, 4)
print('5x4 Matrix:\n', matrix)

# Mean, Median, Std
print('Mean:', np.mean(arr))
print('Median:', np.median(arr))
print('Standard Deviation:', np.std(arr))

# Element-wise operations
arr2 = np.arange(30, 50)
print('Addition:', arr + arr2)
print('Multiplication:', arr * arr2)
```

**Matplot lib**

```python
import matplotlib.pyplot as plt

# Line Plot
x = list(range(1, 8))
y = [i*3 for i in x]   # multiples of 3
plt.plot(x, y, marker='o')
plt.title('Numbers vs Multiples of 3')
plt.xlabel('Numbers')
plt.ylabel('Multiples of 3')
plt.show()

# Bar Chart
students = ['Baloxh', 'Mona', 'Sohaib', 'Zainab', 'Hassan']
marks = [72, 95, 81, 67, 88]
plt.bar(students, marks, color='skyblue')
plt.title('Student Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

# Pie Chart
categories = ['Food', 'Transport', 'Entertainment', 'Savings']
values = [40, 25, 20, 15]
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Monthly Budget Distribution')
plt.show()
```
