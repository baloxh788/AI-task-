```python
# Simple Tic Tac Toe Game (2 Players)

# Step 1: Create the board
board = [" " for _ in range(9)]

# Step 2: Function to display the board
def show_board():
    print("\n")
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print("\n")

# Step 3: Function to check winner
def check_winner(player):
    # All winning combinations
    wins = [
        [0,1,2], [3,4,5], [6,7,8],   # rows
        [0,3,6], [1,4,7], [2,5,8],   # columns
        [0,4,8], [2,4,6]             # diagonals
    ]
    for combo in wins:
        if board[combo[0]] == board[combo[1]] == board[combo[2]] == player:
            return True
    return False

# Step 4: Main game loop
current_player = "X"

for turn in range(9):  # Max 9 moves
    show_board()
    print(f"Player {current_player}'s turn.")

    # Take input
    move = int(input("Choose a position (1-9): ")) - 1

    # Check if valid
    if board[move] != " ":
        print("That spot is taken! Try again.")
        continue

    # Make move
    board[move] = current_player

    # Check winner
    if check_winner(current_player):
        show_board()
        print(f"üéâ Player {current_player} wins!")
        break

    # Switch player
    current_player = "O" if current_player == "X" else "X"
else:
    show_board()
    print("It's a draw! üòê")
```
